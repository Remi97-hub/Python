```python
import pandas as pd
```

```python
dataset=pd.read_csv("MUL REG& DT.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
dataset
```

```python
dataset.columns
```

```python
independant=dataset[['ID', 'AGE', 'EXPERIENCE',
       'EDUCATION_MASTERS', 'EDUCATION_PhD', 
       'CITY_LOS ANGELES', 'CITY_NEW YORK']]
dependant=dataset[['SALARY']]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independant,dependant,test_size=0.20,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python

scy=StandardScaler()
y_train=scy.fit_transform(y_train)
y_test=scy.transform(y_test)
```

```python
from sklearn.svm import SVR
regressor=SVR(kernel='linear',C=1000)
regressor.fit(x_train,y_train)
```

```python
regressor.intercept_
```

```python
regressor.n_support_
```

```python
regressor.support_
```

```python
y_pred=regressor.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
```

```python
r_score
```

```python
import pickle
filename="final_preprocessed_model.sav"
```

```python
pickle.dump(regressor,open(filename,'wb'))
```
