```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
def calc_vif(x):
    vif=pd.DataFrame()
    vif["Variables"]=x.columns
    vif["VIF"]=[variance_inflation_factor(x.values,i) for i in range(x.shape[1])]
    return (vif)
```

```python
calc_vif(a)
```

```python
b=calc_vif(a)
c=b.drop(index=[4,2,3])
```

```python
#T-test
#Unpaird T-test (comparing two variables under one condition)
```

```python
from scipy.stats import ttest_ind
male=dataset[dataset["gender"]=='M']["salary"]
female=dataset[dataset["gender"]=='F']["salary"]
ttest_ind(male,female)
```

```python
#Hypothesis
H0: there is no relation between gender and salary
H1: there is relation between gender and salary
#p-value<0.05 , so there is significant difference (salary of male and female is different)
#stick with H0
```

```python
#Paired T-test
```

```python
from scipy.stats import ttest_rel
female=dataset[dataset["gender"]=='F']["etest_p"]
female1=dataset[dataset["gender"]=='F']["salary"]
ttest_rel(female,female1)
```

```python
ttest_rel(female1,female)
```

```python
#Hypothesis
H0: There is no relation between entrance test mark and salary of female
H1: there is relation between salary and entrance test marks of female
# the p-value>0.05 which tells that there is no significant relation between both
#move to H1
```

```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Fixed the formula string by using proper quoting
model = ols('salary ~ C(degree_t) + C(specialisation) + C(degree_t):C(specialisation)', data=dataset).fit()
anova_table = sm.stats.anova_lm(model, type=2)
```
