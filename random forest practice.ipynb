```python
import pandas as pd
dataset=pd.read_csv("MUL REG& DT.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
dataset
```

```python
dataset.columns
```

```python
independant=dataset[['ID', 'AGE', 'EXPERIENCE','EDUCATION_BACHELORS',
       'EDUCATION_MASTERS', 'EDUCATION_PhD', 'CITY_CHICAGO',
       'CITY_LOS ANGELES', 'CITY_NEW YORK']]
```

```python
dependant=dataset[['SALARY']]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independant,dependant,test_size=0.30,random_state=0)
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor=RandomForestRegressor(n_estimators=200,criterion="poisson",random_state=0)
regressor.fit(x_train,y_train)
```

```python
y_pred=regressor.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
```

```python
r_score
```

```python
import pickle
filename="final_random_forest_model.sav"
```

```python
pickle.dump(regressor,open(filename,'wb'))
```
