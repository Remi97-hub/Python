```python
import pandas as pd
```

```python
dataset=pd.read_csv("MUL REG& DT.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
dataset.drop('ID',axis=1)
```

```python
dataset.columns
```

```python
indep=dataset[['AGE', 'EXPERIENCE','EDUCATION_MASTERS',
       'EDUCATION_PhD', 'CITY_LOS ANGELES', 'CITY_NEW YORK']]
dep=dataset[['SALARY']]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(indep,dep,test_size=0.30,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
import pickle
with open ("all_grid_models.pkl",'rb') as f:loaded_grids=pickle.load(f)
    
```

```python
dt_grid=loaded_grids['dt']
dt_grid.fit(x_train,y_train)
```

```python
re=dt_grid.cv_results_
grid_prediction=dt_grid.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(y_test,grid_prediction)
```

```python
r_score
```

```python
print("The R score value for the best parameter {}:".format(dt_grid.best_params_),r_score)
```
